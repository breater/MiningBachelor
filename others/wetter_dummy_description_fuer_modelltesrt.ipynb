```python
import pandas as pd
```

```python
frame = pd.read_csv('20152018/weatherforWestStChambersSt.csv')

df = pd.DataFrame(frame)
```

```python
df = pd.get_dummies(df, columns=['weekday', 'hour','weatherDesc2','month'] )
```

```python
import holidays 
def getHoliday(year,month,day):
    us_holidays = holidays.US()
    return  1 if(f"{day}-{month}-{year}" in us_holidays) else 0
```

```python
def getholi(data):
    #print(data)
    return getHoliday(data['year'],data['month'],data['day'])
```

```python
import datetime
def getWeekday(year,month,day):    
    return  datetime.date(year,month,day).weekday()
```

```python
def getweekday(data):
    #print(data)
    return getWeekday(data['year'],data['month'],data['day'])
```

```python
def getworkday(weekday,holiday):
    return  1 if( weekday < 5 and holiday == 0) else 0
```

```python
def getworkingday(data):
    #print(data)
    return getworkday(data['weekday'],data['holidays'])
```

```python
df['holidays'] = df.apply(getholi,axis=1)
```

```python
df['weekday'] = df.apply(getweekday,axis=1)
```

```python
df['workday'] = df.apply(getworkingday,axis=1)
```

```python
#Ausleihen mit Predictions laden
frame_ausleihen = pd.read_csv(r'20152018/ausleiheformatzted.csv')

df_ausleihen = pd.DataFrame(frame_ausleihen)

df_ausleihen.head()
```

```python
def getPrediction(year,month,day,hour):
    prediction = df_ausleihen[(df_ausleihen['year'] == year) & (df_ausleihen['month'] == month) & (df_ausleihen['day']==day) & (df_ausleihen['hour']==hour)]
    if prediction['count'].empty == False:
        return prediction['count'].values[0]
    else:
        return 0
```

```python
def getPredict(data):
    return getPrediction(data['year'],data['month'],data['day'],data['hour'])
```

```python
df['ausleihen'] = df.apply(getPredict,axis=1)
```
