```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#frame = pd.read_csv(r'ausleihe2.csv')
frame = pd.read_csv(r'ausleihen_mit_wetter.csv')
df = pd.DataFrame(frame)
col = list(df.columns.values)

test = pd.DataFrame(frame)

df.head()
```

```python
df_2019 = df.drop(['year']==2019, axis=1, errors='ignore')
```

```python
df_2019
```

```python
# Visualisierung
fig, ax = plt.subplots(figsize=(20,10))
sns.pointplot(data=df_2019, x='week',y='count',hue='station1',ax=ax)
```

```python
corr = df_2019.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr,annot=True)
```

```python
fix,(ax1,ax2) = plt.subplots(ncols=2,figsize=(20,6))
sns.regplot(x=df_2019['precipMM'], y=df_2019['count'],ax=ax1)
ax1.set(title="Relation zwischen Temp und Ausleihen")
sns.regplot(x=df_2019['week'], y=df_2019['count'],ax=ax2)
ax2.set(title="Relation zwischen Wochen und Ausleihen")
```
