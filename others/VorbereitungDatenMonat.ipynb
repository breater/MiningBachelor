```python
import pandas as pd
import numpy as np
import datetime
import math
```

```python
#frame = pd.read_csv(r'201905_first.csv')
frame = pd.read_csv(r'201905_modell1_weather.csv')
df = pd.DataFrame(frame)

df.head()
```

```python
# Drop Trips shorter than 90 seconds
df = df.loc[((df["tripduration"]>90)),:]
df.head()
```

```python
# Drop Trips longer than 24 hours
df = df.loc[df["tripduration"] < 1440, :]
df.head()
```

```python
# Drop unused columns
df = df.drop(['hour','end station id','end station id','end station name', 'end station latitude','end station longitude','birth year','gender','start station id','tripduration','usertype','stoptime'], axis=1, errors='ignore')
df.head()
```

```python
df.dtypes
```

```python
#Starttime to date-format
df['starttime'] = pd.to_datetime(df['starttime'], errors='coerce')
```

```python
df['starttime'] = df['starttime'].dt.date
df.head()
```

```python
df['dayofweek'] = df['starttime'].dt.day_of_week
df.head()
```

```python
df['hour'] = df['starttime'].dt.hour
df.head()
```

```python
df
```

```python
df = df.groupby([df['starttime'], df['start station name'],df['hour'],df['start station latitude'],df['start station longitude'],df['dayofweek']])['bikeid'].agg('count')
```

```python
df.to_csv('201905_first.csv')
```

```python
df = pd.DataFrame(df)
```

```python
df.sort_values(by=['starttime','hour','start station name'])
```

```python
df.to_csv('201905_first.csv')
```

```python
def getLastweek(starttime,hour,station,distance):
    
    daybefore = starttime - datetime.timedelta(days=distance)
    #print(daybefore)
    
    value = df[(df['starttime'] == daybefore) & (df['hour']== hour) &  (df['start station']== station)]
    if value['anzahl'].empty == False:
        return value['anzahl'].values[0]
    return math.nan
```

```python
def getWeek1(data):
    print(data)
    return getLastweek(data['starttime'],data['hour'],data['start station'],1)
```

```python
local = df.copy()
# The apply method calls a function on each row
local['day1anzahl'] = local.apply(getWeek1,axis=1)
```

```python
df.shape
```

```python
# wir spielen an df

df = df.fillna(0)
df.head()
```

```python
df.to_csv('201905_first_day1.csv',index=False)
```

```python
def applyFunc(s):
    if s == 5 or s == 6:
        return 0
    else:
        return 1
    return ''
```

```python
df['weekday'] = df['dayofweek'].apply(applyFunc)
```

```python
df.to_csv('201905_day1_and_weekday.csv',index=False)
```
