```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
frame = pd.read_csv(r'2017_2018_ausleihen_count.csv')
#frame = pd.read_csv(r'ausleihen_mit_wetter.csv')
df = pd.DataFrame(frame)
df.head()
```

```python
df
```

```python
df = df.groupby([df['year'],df['month'],df['day'],df['hour'],df['dayofweek'],df['date']])['count1'].agg('sum')
```

```python
df['start'] = pd.to_datetime(df['start'], errors='coerce')
```

```python
df['week'] = df['start'].dt.isocalendar().week
```

```python
df.to_csv('2017_2018_merged_counted_06-09.csv')
```

```python
df_cleaned = df.groupby([df['year'],df['week'],df['dayofweek'],df['hour'],df['month']])['bikeid'].agg('sum')
```

```python
df_cleaned
```

```python
df_cleaned = pd.DataFrame(df_cleaned)
```

```python
df_cleaned.to_csv('2017_2018_merged_counted_06-09.csv')
```

```python
frame_weather = pd.read_csv(r'weather20172018.csv')
#frame = pd.read_csv(r'ausleihen_mit_wetter.csv')
df_weather = pd.DataFrame(frame_weather)
df_weather
```

```python
df_weather['start'] = pd.to_datetime(df_weather['start'], errors='coerce')
```

```python
df_weather['dayofweek'] = df_weather['start'].dt.day_of_week
df_weather['week'] = df_weather['start'].dt.isocalendar().week
```

```python
df_weather['hour'] = df_weather['start'].dt.hour
df_weather['year'] = df_weather['start'].dt.year
df_weather['month'] = df_weather['start'].dt.month
```

```python
df_weather
```

```python
df_weather = df_weather.drop(['start'], axis=1)
```

```python
#merge weather and rentals
```

```python
df_merged = pd.merge(df,df_weather,on=['date','hour'],how='left')

df_merged.head()
```

```python
df_merged
```

```python
df_merged = df_merged.dropna()
```

```python
df_merged.to_csv('2017_2018_weather_merged_06.09.csv',index=False)
```
