```python
import holidays 
def getHoliday(year,month,day):
    us_holidays = holidays.US()
    return  1 if(f"{day}-{month}-{year}" in us_holidays) else 0
```

```python

def getholi(data):
    #print(data)
    return getHoliday(data['year'],data['month'],data['day'])
```

```python
import datetime
def getWeekday(year,month,day):    
    return  datetime.date(year,month,day).weekday()
```

```python
def getweekday(data):
    #print(data)
    return getWeekday(data['year'],data['month'],data['day'])
```

```python
def getworkday(weekday,holiday):
    return  1 if( weekday < 5 and holiday == 0) else 0
```

```python
def getworkingday(data):
    #print(data)
    return getworkday(data['weekday'],data['holidays'])
```

```python
ausleihe['holidays'] = ausleihe.apply(getholi,axis=1)
test['holidays'] = test.apply(getholi,axis=1)
```

```python
ausleihe['weekday'] = ausleihe.apply(getweekday,axis=1)
test['weekday'] = test.apply(getweekday,axis=1)
```

```python
ausleihe['workday'] = ausleihe.apply(getworkingday,axis=1)
test['workday'] = test.apply(getworkingday,axis=1)
```
