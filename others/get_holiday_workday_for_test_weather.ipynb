```python
import pandas as pd
```

```python
frame_alr = pd.read_csv(r'20152018/wetter082018_mit_ausleihen_und_rueckgaben.csv')

df_alr = pd.DataFrame(frame_alr)

df_alr.head()
```

```python
import holidays 
def getHoliday(year,month,day):
    us_holidays = holidays.US()
    return  1 if(f"{day}-{month}-{year}" in us_holidays) else 0
```

```python
def getholi(data):
    #print(data)
    return getHoliday(data['year'],data['month'],data['day'])
```

```python
import datetime
def getWeekday(year,month,day):    
    return  datetime.date(year,month,day).weekday()
```

```python
def getweekday(data):
    #print(data)
    return getWeekday(data['year'],data['month'],data['day'])
```

```python
def getworkday(weekday,holiday):
    return  1 if( weekday < 5 and holiday == 0) else 0
```

```python
def getworkingday(data):
    #print(data)
    return getworkday(data['weekday'],data['holidays'])
```

```python
df_alr['holidays'] = df_alr.apply(getholi,axis=1)
```

```python
df_alr['weekday'] = df_alr.apply(getweekday,axis=1)
```

```python
df_alr['workday'] = df_alr.apply(getworkingday,axis=1)
```
