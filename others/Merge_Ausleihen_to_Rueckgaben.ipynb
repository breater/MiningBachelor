```python
import pandas as pd
```

```python
#Ausleihen mit Predictions laden
frame_ausleihen = pd.read_csv(r'20152018/ruckgabeformatzted.csv')

df_ausleihen = pd.DataFrame(frame_ausleihen)

df_ausleihen.head()
```

```python
#Rueckgaben laden
frame_wetter = pd.read_csv(r'20152018/wetter082018_mit_ausleihen.csv')

df_wetter = pd.DataFrame(frame_wetter)

df_wetter.head()
```

```python
def getPrediction(year,month,day,hour):
    prediction = df_ausleihen[(df_ausleihen['year'] == year) & (df_ausleihen['month'] == month) & (df_ausleihen['day']==day) & (df_ausleihen['hour']==hour)]
    if prediction['count'].empty == False:
        return prediction['count'].values[0]
    else:
        return 0
```

```python
def getPredict(data):
    return getPrediction(data['year'],data['month'],data['day'],data['hour'])
```

```python
df_wetter['rueckgaben'] = df_wetter.apply(getPredict,axis=1)
```

```python
#Weil Zeile überflüssig war bzw. Namen geändert wurde
df_wetter = df_wetter.drop(['prediction'],axis=1)
```

```python
df_wetter.to_csv('20152018/wetter082018_mit_ausleihen_und_rueckgaben.csv',index=False)
```
