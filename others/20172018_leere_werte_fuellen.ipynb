```python
######
import pandas as pd
```

```python
frame = pd.read_csv(r'2017_2018_ausleihen_count.csv')
#frame = pd.read_csv(r'ausleihen_mit_wetter.csv')
df = pd.DataFrame(frame)
```

```python
df['start'] = pd.to_datetime(df['start'], errors='coerce')
```

```python
r = pd.date_range(start=df['start'].min(), end=df['start'].max(), freq='H')
df.set_index('start').reindex(r).fillna(0.0).rename_axis('start').reset_index()
```
