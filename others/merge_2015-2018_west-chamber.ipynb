```python
import pandas as pd
```

```python
frame = pd.read_csv(r'20152018/ruckgabeformatzted.csv')
#frame = pd.read_csv(r'ausleihen_mit_wetter.csv')
df_ausleihe = pd.DataFrame(frame)
df_ausleihe.head()
```

```python
frame_weather = pd.read_csv(r'20152018/weatherforWestStChambersSt.csv')
#frame = pd.read_csv(r'ausleihen_mit_wetter.csv')
df_weather = pd.DataFrame(frame_weather)
df_weather.head()
```

```python
df_neu = pd.merge(df_ausleihe,df_weather,on=['year','month','day','hour'],how='left')

df_neu.head()
```
